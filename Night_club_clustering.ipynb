```python
#Combine Fairness, WOM, Revisit, Familiarity Score from their separate scores
club_data['fair_score'] = club_data[['Fair1', 'Fair2', 'Fair3']].mean(axis=1)
club_data['wom_score'] = club_data[['WOM_1', 'WOM_2', 'WOM_3']].mean(axis=1)
club_data['ri_score'] = club_data[['RI_1', 'RI_2', 'RI_3']].mean(axis=1)
club_data['fam_score'] = club_data[['FAM1', 'FAM2']].mean(axis=1)
```

```python
#Now delete the uncombined Variables
club_data = club_data.drop(columns=['Fair1', 'Fair2', 'Fair3','WOM_1', 'WOM_2', 'WOM_3','RI_1', 'RI_2', 'RI_3','FAM1', 'FAM2'])
```

```python
#Combine 'Daily' and '2-3 Times a Week' in  Nightclub_Frequency
club_data['nightclub_frequency'] = club_data['Nightclub_Frequency'].replace('Daily', '2-3 Times a Week')
club_data = club_data.drop(columns=['Nightclub_Frequency'])
```

```python
# Make Age categorical 'under 25', 'under 45','45+'
club_data['age'] = np.where(club_data['Age'] < 25, 'under 25',
                    np.where(club_data['Age'] < 45, '25 - 45', '45+'))
club_data = club_data.drop(columns=['Age'])
```

```python
#drop 'Ethnicity', 'Employment_status' and 'Education
club_data = club_data.drop(columns=['Ethnicity','Education','Employment_status'])
```

```python
#rename all columns to lowercase style
club_data.columns = club_data.columns.str.lower()
```

##Clustering

####K-prototype clustering

```python
!pip install kmodes
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
# Get indices of categorical columns (needed for KPrototypes)
cat_indices = [club_data.columns.get_loc(col) for col in cat_cols]
```

```python
# Testing  different K values
k_values = range(2, 11)
costs = []
silhouettes = []
```

```python
print("Testing different K values...")
for k in k_values:
    kproto = KPrototypes(n_clusters=k, init='Cao', random_state=42, verbose=0)
    clusters = kproto.fit_predict(club_data, categorical=cat_indices)

    # Cost (lower is better)
    cost = kproto.cost_
    costs.append(cost)

    print(f"K={k}: Cost={cost:.2f}, Iterations={kproto.n_iter_}")
```

```python
import matplotlib.pyplot as plt
```

```python
# Plotting the elbow curve
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(k_values, costs, 'bo-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Cost', fontsize=12)
plt.title('Elbow Method for Optimal K', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(k_values)
```

```python
# Calculate rate of decrease (to help identify elbow)
plt.subplot(1, 2, 2)
rate_of_decrease = [0] + [costs[i-1] - costs[i] for i in range(1, len(costs))]
plt.plot(k_values, rate_of_decrease, 'ro-', linewidth=2, markersize=8)
plt.xlabel('Number of Clusters (K)', fontsize=12)
plt.ylabel('Cost Decrease', fontsize=12)
plt.title('Rate of Cost Decrease', fontsize=14)
plt.grid(True, alpha=0.3)
plt.xticks(k_values)

plt.tight_layout()
plt.show()
```

Optimal k is when k=5

```python
#the dataset
data = club_data.copy()
```

```python
# Fit final model with chosen K
optimal_k = 7

kproto_final = KPrototypes(n_clusters=optimal_k, init='Cao', random_state=42)
clusters = kproto_final.fit_predict(data, categorical=cat_indices)
```

```python
# Add cluster labels to data
data['cluster'] = clusters
```

```python
# Show cluster statistics
print(f"\nCluster sizes:")
print(data['cluster'].value_counts().sort_index())
print('---------------------------------------------')
print(f"\nCluster centers:")
for i, centroid in enumerate(kproto_final.cluster_centroids_):
    print(f"\nCluster {i}:")
    for j, col in enumerate(num_cols+cat_cols):
        cat_idx = j
        print(f"  {col}: {centroid[cat_idx]}")
```

###Agglomerative Clustering

```python
#we will use Gower distance
!pip install gower
```

```python
import gower
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
import seaborn as sns
```

```python
agglomerative_data = club_data.copy()
```

```python
#Calculate Gower distance matrix
distance_matrix = gower.gower_matrix(agglomerative_data)
```

```python
#Convert to condensed distance matrix for scipy
# scipy needs a condensed format (upper triangle only)
distance_condensed = squareform(distance_matrix, checks=False)
```

```python
# Perform hierarchical clustering
# Linkage methods:  'average'
linkage_matrix = linkage(distance_condensed, method='complete')
```

```python
#Create dendrogram
plt.figure(figsize=(15, 7))
dendrogram(linkage_matrix,
           truncate_mode='lastp',  # Show only last p merged clusters
           p=30,  # Show last 30 merges
           leaf_rotation=90,
           leaf_font_size=10,
           show_contracted=True)
plt.title('Hierarchical Clustering Dendrogram (Agglomerative)', fontsize=16)
plt.xlabel('Cluster Size', fontsize=12)
plt.ylabel('Gower Distance', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Cut the dendrogram to form clusters
n_clusters = 7
clusters = fcluster(linkage_matrix, t=n_clusters, criterion='maxclust')

# Add clusters to dataframe
agglomerative_data['Hierarchical_Cluster'] = clusters

print(f"\nCluster distribution:")
print(agglomerative_data['Hierarchical_Cluster'].value_counts().sort_index())
```

Note that, I first used Average linkage and the clusters unevenly distributed

```python
  # Numerical features summary
numerical_summary = agglomerative_data.groupby('Hierarchical_Cluster')[num_cols].agg([
  'mean', 'median', 'std', 'count'
]).round(2)

numerical_summary
```

```python
# Get modes for cat data in the cluster
cat_modes = agglomerative_data.groupby('Hierarchical_Cluster')[cat_cols].agg(
    lambda x: x.mode()[0] if not x.mode().empty else None
)

cat_modes
```

##Interpretation

###K-Prototypes Clustering

Key distinctions across clusters include perceived pricing fairness, intent to revisit, and demographics:

Cluster 0 – Young Loyal Students

Female, student, low-income frequent customers using VIP entrances. Although they rate fairness below average (−0.63), they show strong word-of-mouth (WOM ≈ 0.53) and revisit intent (0.63). These are emotionally attached regulars who stay loyal despite perceiving unfair pricing.

Cluster 1 – Older First-Timers

Male, older first-time customers attracted by “day-of-week” pricing. Their fairness perception is slightly below average (−0.37), yet WOM and revisit scores remain moderate, suggesting curiosity and openness to return if value is communicated clearly.

Cluster 2 – Wealthy Value-Seekers

High-income, full-time employed females who view pricing as highly fair (+1.32) but visit infrequently. They approve of the structure but are not deeply engaged with the nightclub experience.

Cluster 3 – Dissatisfied Regulars

Mid-income female students who visit 2–3 times monthly. They pay flat pricing and show the lowest fairness and revisit scores, signaling frustration among loyal patrons who feel pricing lacks transparency.

Cluster 4 – Affluent Critics

High-income employed females reserving in advance. They perceive fairness positively (+1.24) but have the most negative WOM (−1.47) and revisit intent (−1.49), indicating intellectual satisfaction but emotional disengagement.


Overall, fairness perception and revisit intent are the main drivers separating clusters: frequent students feel pricing is unfair but stay loyal, whereas affluent patrons feel pricing is fair but remain detached.

###Agglomerative (Hierarchical) Clustering

Interpretation:

Clusters 2 and 3 exhibit the most positive fairness and revisit scores, representing satisfied, high-value customers.

Cluster 1 reflects the most dissatisfied patrons—negative across fairness, WOM, and revisit, signaling frustration with flat pricing.

Cluster 4 is moderate in all dimensions, representing the average satisfied frequent visitor.



Both methods consistently highlight fairness perception as the strongest differentiator and confirm the coexistence of loyal but critical frequent students and satisfied yet detached high-income customers.

##Insight


Frequent Students (K-Proto 0 & 3; Hier 4): Clarify and justify pricing rules. Offer student-night discounts or loyalty tiers to rebuild perceived fairness.

First-Timers (K-Proto 1; Hier 3): Maintain promotional weekday pricing; use follow-up offers to convert them into regulars.

Reward referrals from satisfied segments (Hier 2 & 3; K-Proto 0 & 2).

Collect feedback from dissatisfied clusters (Hier 1; K-Proto 3) to fix perceived unfairness and improve the customer journey.
